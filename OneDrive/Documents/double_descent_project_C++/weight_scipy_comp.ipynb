```python
df = pd.read_csv("daily_vix.csv")
close = df["Close"].values
```

```python
ret = lambda x: (x[1:]- x[:-1])/ x[:-1]
log_ret = lambda x: np.log(x[1:]/ x[:-1])

#lag matrix creation
lag_matrix = lambda x,lag: np.array([[x[j+i] for j in range(lag)] for i in range(len(x)-lag)])
```

```python
n_lags = 14
ret_price = log_ret(close)
close_lag_mat = lag_matrix(ret_price, n_lags)
```
