```python
c = 0.99894330049030156
s = 0.045959573600533918
val_1 = mat[0,0]
val_2 = mat[0, 2]
neg_func_1 = lambda c, s, val_1, val_2: val_1 * c - val_2 * s
pos_func_2 = lambda c, s, val_1, val_2: val_2 * c + val_1 * s
neg_func_1(c, s, val_1, val_2), pos_func_2(c, s, val_1, val_2)
```

```python
for i in range(4):
    val_1 = mat[i,0]
    val_2 = mat[i, 2]
    temp_1 = neg_func_1(c, s, val_1, val_2)
    temp_2 = pos_func_2(c, s, val_1, val_2)
    print(temp_1, temp_2)
```
