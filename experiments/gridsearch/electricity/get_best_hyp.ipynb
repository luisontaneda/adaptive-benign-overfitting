```python
import pandas as pd

df = pd.read_csv('../../results/gridsearch/mse_var_grid_1.csv')
```

```python
g_abo = (
    df.groupby(["L", "W"], as_index=False)
      .agg(
          mse_mean=("mse_abo", "mean"),
          var_mean=("var_abo", "mean")
      )
)
best_abo_mse = g_abo.loc[g_abo["mse_mean"].idxmin()]
best_abo_mse
```
